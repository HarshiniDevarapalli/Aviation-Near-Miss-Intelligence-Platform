```python
from google.colab import files
import pandas as pd
import glob

# Upload all 7 CSV files
uploaded = files.upload()

# Read and merge all CSVs
dfs = []
for file in glob.glob("*.csv"):
    dfs.append(pd.read_csv(file))

merged_df = pd.concat(dfs, ignore_index=True)

print("Merged Shape:", merged_df.shape)
merged_df.head()
```

```python
# NASA exports usually have the first row as actual headers

new_header = merged_df.iloc[0]

merged_df = merged_df[1:]
merged_df.columns = new_header
merged_df.reset_index(drop=True, inplace=True)

print("Columns:\n")
for col in merged_df.columns:
    print(col)

print("\nDataset Shape:", merged_df.shape)

merged_df.head()
```

```python
import pandas as pd

# Rename duplicate narrative columns
cols = list(merged_df.columns)

narrative_count = 0
new_cols = []

for col in cols:
    if col == "Narrative":
        narrative_count += 1
        new_cols.append(f"Narrative_{narrative_count}")
    else:
        new_cols.append(col)

merged_df.columns = new_cols

# Create combined narrative
merged_df["full_narrative"] = (
    merged_df["Narrative_1"].fillna("") +
    " " +
    merged_df["Narrative_2"].fillna("")
)

# Keep only useful columns
clean_df = merged_df[
    [
        "ACN",
        "Date",
        "Primary Problem",
        "Contributing Factors / Situations",
        "Result",
        "Synopsis",
        "full_narrative"
    ]
]

# Remove empty reports
clean_df = clean_df.dropna(subset=["full_narrative"])

clean_df = clean_df[
    clean_df["full_narrative"].str.strip().str.len() > 50
]

clean_df.reset_index(drop=True, inplace=True)

print("Final Shape:", clean_df.shape)

clean_df.head()
```

```python
clean_df.to_csv(
    "asrs_clean.csv",
    index=False
)

print("Dataset saved!")
```
