```python
from sentence_transformers import SentenceTransformer
import chromadb

print("Everything installed successfully!")
```

```python
import pandas as pd

df = pd.read_csv("asrs_clean.csv")

print(df.shape)
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(df["full_narrative"].duplicated().sum())
```

```python
df["length"] = df["full_narrative"].str.len()

print(df["length"].describe())
```

```python
df = df.dropna(subset=["full_narrative"])

df = df[
    df["full_narrative"].str.len() > 100
]

df = df.drop_duplicates(
    subset=["full_narrative"]
)

print(df.shape)
```

```python
df.to_csv(
    "asrs_clean_v2.csv",
    index=False
)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "BAAI/bge-small-en-v1.5"
)

print("Model Loaded")
```

```python
documents = (
    "Primary Problem: " + df["Primary Problem"].fillna("") +
    "\nContributing Factors: " + df["Contributing Factors / Situations"].fillna("") +
    "\nResult: " + df["Result"].fillna("") +
    "\nSynopsis: " + df["Synopsis"].fillna("") +
    "\nNarrative: " + df["full_narrative"].fillna("")
).tolist()

print(len(documents))
```

```python
embeddings = model.encode(
    documents,
    show_progress_bar=True
)
```

```python
print(embeddings.shape)
```

```python
import numpy as np

np.save(
    "asrs_embeddings.npy",
    embeddings
)

print("Embeddings saved!")
```

```python
df.to_csv(
    "asrs_clean_final.csv",
    index=False
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

query = """
Aircraft taxiing in low visibility.
Communication delays reported.
"""

query_embedding = model.encode([query])

scores = cosine_similarity(
    query_embedding,
    embeddings
)[0]

top_indices = scores.argsort()[-5:][::-1]
```

```python
for idx in top_indices:
    print("="*80)
    print(df.iloc[idx]["full_narrative"][:1000])
    print("\nScore:", scores[idx])
```

```python
import chromadb

client = chromadb.Client()

collection = client.create_collection(
    name="aviation_incidents"
)

print("Collection Created")
```

```python
documents = df["full_narrative"].tolist()
```

```python
batch_size = 5000

for i in range(0, len(documents), batch_size):

    collection.add(
        ids=[str(j) for j in range(i, min(i + batch_size, len(documents)))],
        documents=documents[i:i + batch_size],
        embeddings=embeddings[i:i + batch_size].tolist()
    )

    print(f"Added batch {i//batch_size + 1}")

print("All documents stored successfully!")
```

```python
import chromadb

client = chromadb.PersistentClient(
    path="./chroma_db"
)

collection = client.get_or_create_collection(
    name="aviation_incidents"
)

print("Persistent ChromaDB Created!")
```

```python
batch_size = 5000

for i in range(0, len(documents), batch_size):

    collection.add(
        ids=[str(j) for j in range(i, min(i + batch_size, len(documents)))],
        documents=documents[i:i + batch_size],
        embeddings=embeddings[i:i + batch_size].tolist()
    )

    print(f"Added batch {i//batch_size + 1}")

print("Done!")
```
